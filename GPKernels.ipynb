```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process.kernels import RBF, ExpSineSquared, RationalQuadratic, WhiteKernel, Matern
from sklearn.gaussian_process.kernels import Kernel
```

```python
def plot_covariance_matrix(kernel: Kernel, X: np.ndarray, ax: plt.Axes, title: str, cmap: str, vmin: float, vmax: float):
    """Compute and plot the covariance matrix for a given kernel."""
    K = kernel(X)
    cax = ax.matshow(K, cmap=cmap, vmin=vmin, vmax=vmax)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.colorbar(cax, ax=ax, shrink=0.75, pad=0.02)

# Define the range of X
X = np.linspace(-3, 3, 100).reshape(-1, 1)

# Create subplots for each kernel
fig, axes = plt.subplots(1, 4, figsize=(22, 5))
plt.subplots_adjust(wspace=0.3)

# Define kernels with unique color maps
kernels = [
    (RBF(length_scale=1.0), 'RBF Kernel', 'plasma'),
    (RationalQuadratic(alpha=1.0, length_scale=1.0), 'RationalQuadratic Kernel', 'viridis'),
    (Matern(length_scale=1.0, nu=1.5), 'Matern Kernel (nu=1.5)', 'inferno'),
    (WhiteKernel(noise_level=1.0), 'WhiteKernel', 'coolwarm')
]

# Calculate covariance matrices to determine the common color scale
all_matrices = []
for kernel, _, _ in kernels:
    K = kernel(X)
    all_matrices.append(K)

# Find global min and max values for consistent color scaling
all_matrices = np.concatenate([K.flatten() for K in all_matrices])
vmin, vmax = all_matrices.min(), all_matrices.max()

# Plot each kernel's covariance matrix with distinct color maps
labels = ['(a)', '(b)', '(c)', '(d)']
for i, (kernel, title, cmap) in enumerate(kernels):
    plot_covariance_matrix(kernel, X, axes[i], title, cmap, vmin, vmax)
    axes[i].text(0.5, 0.1, labels[i], transform=axes[i].transAxes, fontsize=14, fontweight='bold', va='top', ha='center')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process.kernels import RBF, RationalQuadratic, Matern, WhiteKernel
from sklearn.gaussian_process.kernels import Kernel


def plot_covariance_matrix(kernel: Kernel,
                           X: np.ndarray,
                           ax: plt.Axes,
                           title: str,
                           cmap: str,
                           vmin: float,
                           vmax: float):
    """Compute and plot the covariance matrix for a given kernel (square shape)."""

    K = kernel(X)

    im = ax.imshow(
        K,
        cmap=cmap,
        vmin=vmin,
        vmax=vmax,
        aspect='equal',         # Forces square pixels
        interpolation='nearest',
        origin='lower'
    )

    ax.set_title(title, fontsize=18, fontweight='bold', pad=15)
    ax.set_xticks([])
    ax.set_yticks([])

    # Make axes square
    ax.set_box_aspect(1)

    # Add colorbar
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)


# -----------------------------
# Define input range
# -----------------------------
X = np.linspace(-3, 3, 100).reshape(-1, 1)

# -----------------------------
# Define kernels
# -----------------------------
kernels = [
    (RBF(length_scale=1.0), 'RBF Kernel', 'plasma'),
    (RationalQuadratic(alpha=1.0, length_scale=1.0), 'Rational Quadratic Kernel', 'viridis'),
    (Matern(length_scale=1.0, nu=1.5), 'Matern Kernel (ν=1.5)', 'inferno'),
    (WhiteKernel(noise_level=1.0), 'White Kernel', 'coolwarm')
]

# -----------------------------
# Compute global color scale
# -----------------------------
all_values = []
for kernel, _, _ in kernels:
    K = kernel(X)
    all_values.append(K.flatten())

all_values = np.concatenate(all_values)
vmin, vmax = all_values.min(), all_values.max()

# -----------------------------
# Create square subplot layout
# -----------------------------
fig, axes = plt.subplots(1, 4, figsize=(20, 6), dpi=300)

labels = ['(A)', '(B)', '(C)', '(D)']

# -----------------------------
# Plot kernels
# -----------------------------
for i, (kernel, title, cmap) in enumerate(kernels):
    plot_covariance_matrix(kernel, X, axes[i], title, cmap, vmin, vmax)

    axes[i].text(
        0.5, -0.12,
        labels[i],
        transform=axes[i].transAxes,
        fontsize=18,
        fontweight='bold',
        va='top',
        ha='center'
    )

plt.tight_layout()

# -----------------------------
# Save high-quality outputs
# -----------------------------
plt.savefig("Figure_3_BasicKernels.pdf", dpi=300, bbox_inches='tight')
plt.savefig("Figure_3_BasicKernels.png", dpi=300, bbox_inches='tight')
plt.savefig("Figure_3_BasicKernels.tiff", dpi=300, bbox_inches='tight')
plt.savefig(
    "Figure_3_BasicKernels.tiff",
    dpi=300,
    bbox_inches='tight',
    pil_kwargs={"compression": "tiff_lzw"}
)

plt.show()
```
